```python
# now we covert text data into numerical data by using label encoder



from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df["type"] = enc.fit_transform(df.type)
df['nameOrig'] = enc.fit_transform(df['nameOrig'])
df['nameDest'] = enc.fit_transform(df['nameDest'])
```

```python

#drop the target data from colunm and copy to other variable

X = df.drop(['isFraud'], axis = 1)
y = df['isFraud']
```

```python
# now split the data X_train,y_train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , random_state =3, test_size = 0.3)
```

```python
# now we create the model  and check the accuracy
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report, f1_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score , plot_roc_curve
from sklearn.ensemble import GradientBoostingClassifier, StackingClassifier



gb = GradientBoostingClassifier()
model=gb
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

f1score = metrics.f1_score(y_test, y_pred).round(3)
accuracy = metrics.classification_report(y_test, y_pred)
```
